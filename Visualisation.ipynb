```python
# import libraries 
from IPython.core.display import HTML
import requests
import json
import d3_lib
```

### get data from the planteome API

```python
r = requests.get('http://dev.planteome.org/api/entity/term/TO:0000326')
response = r.json()

data = json.loads(response["data"]["topology_graph_json"])
del data['meta'] 
```

```python
##data
```

### force graph
zoom + drag + colors

```python
# visualize as force-directed graph in D3
HTML(d3_lib.set_styles(['force_directed_graph']) + 
'<script src="lib/d3/d3.min.js"></script>' + 
      d3_lib.draw_graph('force_directed_graph', {'data': data}) )
```

### Tree
note: D3 tree does not allow multiple parenting.

```python
# create the file
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```

```python
# visualize in D3
HTML(d3_lib.set_styles(['dndTree']) + 
'<script src="lib/d3/d3.min.js"></script>' + 
'<script src="http://code.jquery.com/jquery-1.10.2.min.js"></script>' + 
'<script src="js/dndTree.js"></script>' + 
      d3_lib.draw_graph('dndTree', {'data': data}) )
```
